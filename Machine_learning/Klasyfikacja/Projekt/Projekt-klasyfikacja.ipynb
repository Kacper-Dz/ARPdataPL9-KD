```python
print('Unique values for columns: \n')
print(f'BusinessTravel: {data.BusinessTravel.unique()} \n')
print(f'Department: {data.Department.unique()}\n')
print(f'EducationField: {data.EducationField.unique()}\n')
print(f'Gender: {data.Gender.unique()}\n')
print(f'JobRole: {data.JobRole.unique()}\n')
print(f'MaritalStatus: {data.MaritalStatus.unique()}')
print(f'OverTime: {data.OverTime.unique()}')
print(f'Over18: {data.Over18.unique()}')
print(f'Attrition: {data.Attrition.unique()}')
```

```python
data = data.drop(columns='Over18')
print(data.keys())
```

```python
data['OverTime'] = data['OverTime'].replace({'Yes': 1, 'No': 0})
data['Attrition'] = data['Attrition'].replace({'Yes': 1, 'No': 0})
data['MaritalStatus'] = data['MaritalStatus'].replace({'Single': 0, 'Married': 1, 'Divorced': 2})
data['Gender'] = data['Gender'].replace({'Male': 1, 'Female': 0})
data.rename(columns={"Gender": "IsMale"})
data['BusinessTravel'] = data['BusinessTravel'].replace({'Non-Travel': 0, 'Travel_Rarely': 1, 'Travel_Frequently': 2})
data['Department'] = data['Department'].replace({'Sales': 0, 'Human Resources': 1, 'Research & Development': 2})
data['EducationField']=data['EducationField'].replace({'Life Sciences': 0,
                                                       'Medical':1,
                                                       'Other':2,
                                                       'Technical Degree':3,
                                                       'Marketing':4,
                                                       'Human Resources':5})
Jobdict = {'Laboratory Technician': 0,
           'Manufacturing Director': 1,
           'Research Scientist': 2,
           'Human Resources': 3,
           'Research Director': 4,
           'Sales Representative':5, 
           'Manager': 6,
           'Healthcare Representative': 7,
           'Sales Executive': 8}

data['JobRole'] = data['JobRole'].map(Jobdict)
```
