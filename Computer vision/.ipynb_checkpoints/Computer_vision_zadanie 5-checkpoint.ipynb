Zadanie 5. Rozmiar outputu warstwy konwolucyjnej.
1. Napisz funkcję, która przyjmuje szerokość wejścia, rozmiar filtru, padding, stride i zwraca szerokość wyjścia.
2. Dla stałych pozostałych wartości, sprawdź, jak na rozmiar wyjścia wpływa:
 - rozmiar wejścia
 - rozmiar filtru
 - padding
 - stride
 (np. większy rozmiar wejścia - większy rozmiar wyjścia)
3. *Przygotuj wykresy obrazujące te zależności.

```python
import matplotlib.pyplot as plt

def calculate_output_width(input_width, filter_size, padding, stride):
    return ((input_width - filter_size + 2 * padding) // stride) + 1

def analyze_output_width(input_width, filter_sizes, paddings, strides):
    output_widths = []

    for filter_size in filter_sizes:
        for padding in paddings:
            for stride in strides:
                output_width = calculate_output_width(input_width, filter_size, padding, stride)
                output_widths.append(output_width)

    return output_widths

# Stałe parametry
input_width = 28
filter_sizes = [3, 5, 7]
paddings = [0, 1, 2]
strides = [1, 2, 3]

# Analiza wpływu parametrów na szerokość wyjścia
output_widths = analyze_output_width(input_width, filter_sizes, paddings, strides)

# Wyświetlanie wyników
print("Input Width:", input_width)
print("Filter Sizes:", filter_sizes)
print("Paddings:", paddings)
print("Strides:", strides)
print("Output Widths:", output_widths)

# Wykresy zależności

# Zależność: Rozmiar wejścia vs. Szerokość wyjścia
plt.figure(figsize=(8, 6))
plt.plot([input_width] * len(output_widths), output_widths, 'o')
plt.xlabel("Input Width")
plt.ylabel("Output Width")
plt.title("Input Width vs. Output Width")
plt.xticks([input_width])
plt.grid(True)
plt.show()

# Zależność: Rozmiar filtru vs. Szerokość wyjścia
plt.figure(figsize=(8, 6))
plt.plot(filter_sizes * len(paddings) * len(strides), output_widths, 'o')
plt.xlabel("Filter Size")
plt.ylabel("Output Width")
plt.title("Filter Size vs. Output Width")
plt.xticks(filter_sizes)
plt.grid(True)
plt.show()

# Zależność: Padding vs. Szerokość wyjścia
plt.figure(figsize=(8, 6))
plt.plot(paddings * len(filter_sizes) * len(strides), output_widths, 'o')
plt.xlabel("Padding")
plt.ylabel("Output Width")
plt.title("Padding vs. Output Width")
plt.xticks(paddings)
plt.grid(True)
plt.show()

# Zależność: Stride vs. Szerokość wyjścia
plt.figure(figsize=(8, 6))
plt.plot(strides * len(filter_sizes) * len(paddings), output_widths, 'o')
plt.xlabel("Stride")
plt.ylabel("Output Width")
plt.title("Stride vs. Output Width")
plt.xticks(strides)
plt.grid(True)
plt.show()
```
