```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cv2_imshow(img): # Funkcja dla jupytera
    img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
    plt.imshow(img)
    plt.show()
```

```python
# Wczytanie obrazu
image_path = 'ML-datasets/images/gory_lc.png'
image = cv.imread(image_path)
cv2_imshow(image)
image.shape
```

```python
# Zmiana rozdzielczości na 256x256
resized_image_1 = cv.resize(image, (256, 256))
cv2_imshow(resized_image_1)
```

```python
# Zmiana rozdzielczości na 40%
scale_percent = 40 # percent of original size

width = int(image.shape[1] * scale_percent / 100)
height = int(image.shape[0] * scale_percent / 100)

resized_image_2_1 = cv.resize(image, (width, height))
cv2_imshow(resized_image_2_1)
```

```python
# Zmiana rozdzielczości na szerokość 256 bez zniekształceń
new_width = 256
new_height = int(image.shape[0] * (new_width / image.shape[1]))
resized_image_3_1 = cv.resize(image, (new_width, new_height))
cv2_imshow(resized_image_3_1)
```

```python
# Korzystajac z funkcji wbudowanej:
resized_image_3_2 = cv.resize(image, fx= 0.4, fy= 0.4, dsize=None)
cv2_imshow(resized_image_3_2)
```

```python
# Matematyczny sposób na znalezienie zadanej szerokości

new_width = 256
pct = new_width / image.shape[1] #liczymy skale dla szerokosci

new_height = int(image.shape[0] * pct)

new_dim = (new_width, new_height)
```

```python
resized_image_3_3 = cv.resize(image, new_dim, interpolation = cv.INTER_AREA)
cv2_imshow(resized_image_3_3)
```
